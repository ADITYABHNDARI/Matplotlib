```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
x = [0,1,2,3,4]

y = [0,2,4,6,8]
# Resize
plt.figure(figsize =(4,3), dpi = 200)

#plt.plot(x,y, label = 'test' , color='r', linewidth = 2, linestyle = '--', marker ='.', markersize='10', markeredgecolor = 'b')

#use shorthand notation
#fmt = '[color][marker][line]'

plt.plot(x,y, 'y^--' , label="test")

## Line Number Two
x2 = np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:6], x2[:6]**2 , 'r', label = 'x^2')
plt.plot(x2[5:], x2[5:]**2 , 'r--')
 

plt.title("Basic", fontdict ={'fontname': 'comic sans MS','fontsize': 20})
plt.xlabel("x axis")
plt.ylabel("y axis")

plt.xticks([1,2,3,4])
#plt.yticks([0,2,4,6,8,10])


plt.legend()

#saving the graph 
plt.savefig('mygraph1.jpg', dpi = 300)

plt.show()
```

### Bar Charts

```python
labels = ['A','B','C']
values = [10,21,13]

bars = plt.bar(labels,values)

patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

#second way
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.figure(figsize= (6,4))

plt.savefig('bar.jpg')

plt.show()
```

### Load FiFa data

```python
fifa = pd.read_csv('fifa.csv')
fifa.head(11)
```

 #### Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins, color = '#abcdef')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.show()
```

#### Pie charts

```python
fifa['Preferred Foot']
```

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

right
```

```python
labels = ['right','left']
colors = ['white', 'violet']

plt.pie([left,right], labels = labels,colors = colors, autopct = '%.2f %%'   )

plt.title('Foot Prefrence of FIFA Players')

plt.show()
```

```python
fifa.Weight
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
plt.style.use('ggplot')
plt.figure(figsize= (7,9))

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa [(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]

weights = [light, light_medium,medium,medium_heavy,heavy]
labels = ['under 125','125-150','150-175','175-200','over 200']
explode = (.4,.2,0,0,.4)
plt.title('Weight Distribution of FIFA Players (in lbs)')

plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance = 0.8, explode=explode)
plt.show()
```

```python
plt.style.use('default')
plt.figure(figsize=[7,10])

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
#rev = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
juv = fifa.loc[fifa.Club == 'Juventus']['Overall']
mu = fifa.loc[fifa.Club == 'Manchester United']['Overall']
chls = fifa.loc[fifa.Club == 'Chelsea']['Overall']



labels = ['FC Barcelona','Real Madrid','Juventus','Manchester United','Chelsea'] #put 'New England Revolution'

boxes = plt.boxplot([barcelona,madrid,juv,mu,chls], labels=labels, patch_artist=True, medianprops={'linewidth':2}) #put rev
for box in boxes['boxes']:
    # set_edge_color
    box.set(color='#4286f4', linewidth = 2)
    # set Fill Color
    box.set(facecolor='#e0e0e0')
    
plt.title('Pro. Soccer team Comparison')
plt.ylabel('FiFa overall Rating')

plt.show()
```
